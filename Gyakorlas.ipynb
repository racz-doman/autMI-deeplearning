```python
import pandas as pd  # konvenció szerint pd aliast használunk
%matplotlib inline
import matplotlib
import numpy as np
```

```python
tweets = []
with open("testset-taska.tsv", encoding="latin1") as f:
    for line in f:
        line = line.split("\t")
        tweet = line[1].split()
        tweet = tweet[0:len(tweet)-1]
        data_tweet = " ".join(tweet)
        tweets.append(data_tweet)
tweets
```

```python
from collections import defaultdict
```

```python
#Szavak számoláas
def word_freq(s):
    freq = defaultdict(int)
    for word in s.split():
        freq[word.lower()] += 1
    return freq

word_freq("black box green tea")
```

```python
words_freq = word_freq(" ".join(tweets))
```

```python
sorted_words = sorted(words_freq.items(), key=lambda kv: kv[1])
```

```python
sorted_words.reverse()
```
